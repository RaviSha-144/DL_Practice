# Learning by example: XOR

If we want to build a neural network in PyTorch, we could specify all our parameters (weight matrices, bias vectors) using Tensors (with requires_grad=True), ask PyTorch to calculate the gradients and then adjust the parameters. In PyTorch, the torch.nn package makes this task of building neural networks more convenient.

The XOR example became famous by the fact that a single neuron, i.e. a linear classifier, cannot learn this simple function, but a network of neurons can.

Creating an NN with one input layer (2 inputs), one hidden layer of tanh activations (4 neurons), and one output layer (1 output).

```python
class simpleClassifier(nn.Module): # nn.Module is the base class for all neural network modules in PyTorch and automatically handles learnable parameters
    def __init__(self, input_size, hidden_size, num_classes):
        super(simpleClassifier, self).__init__() # calls the constructor of the parent class
        self.fc1 = nn.Linear(input_size, hidden_size) # fully connected layer
        self.relu = nn.Tanh() # activation function
        self.fc2 = nn.Linear(hidden_size, num_classes) # fully connected layer

    def forward(self, x): # forward has to be implemented for every nn.Module, backpropagation is automatically handled by PyTorch
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out
    
```

```python
myModel = simpleClassifier(input_size=2, hidden_size=4, num_classes=1)
# Printing a module shows all its submodules
print(myModel)

```

```python
# Obtaining the parameters of a module
params = list(myModel.parameters())
print(params)
print(len(params))
print(params[0].size())  # weights of the first layer
print(params[1].size())  # biases of the first layer
print(params[2].size())  # weights of the second layer
print(params[3].size())  # biases of the second layer

for name, param in myModel.named_parameters():
    print(f"Parameter {name}, shape {param.shape}")

```

Each linear layer has a weight matrix of the shape [output, input], and a bias of the shape [output]. The tanh activation function does not have any parameters. Note that parameters are only registered for nn.Module objects that are direct object attributes, i.e. self.a = .... If you define a list of modules, the parameters of those are not registered for the outer module and can cause some issues when you try to optimize your module. There are alternatives, like nn.ModuleList, nn.ModuleDict and nn.Sequential, that allow you to have different data structures of modules. 

The data package defines two classes which are the standard interface for handling data in PyTorch: data.Dataset, and data.DataLoader. The dataset class provides an uniform interface to access the training/test data, while the data loader makes sure to efficiently load and stack the data points from the dataset into batches during training.

The dataset class summarizes the basic functionality of a dataset in a natural way. To define a dataset in PyTorch, we simply specify two functions: "\_\_getitem\_\_" and "\_\_len\_\_". The get-item function has to return the -th data point in the dataset, while the len function returns the size of the dataset.

```python
class XORdataset(data.Dataset):
    def __init__(self, size: int, std_dev: float = 0.1):
        '''
        Initializes the XOR dataset based on the number of samples and the standard deviation of the noise.
        Inputs:
            size: number of samples
            std_dev: standard deviation of the noise
        '''
        self.size = size
        self.std_dev = std_dev
        self.generate_continuous_XOR()
        
    def generate_continuous_XOR(self):
        '''
        Generates the XOR dataset with continuous values.           
        '''
        data = torch.randint(low=0, high=2, size=(self.size, 2), dtype=torch.float32) # Randomly generates binary data
        label = (data.sum(dim=1) == 1).to(torch.float32) # XOR operation 
        data += self.std_dev * torch.randn(data.shape) # Adds noise to the data
        
        self.data = data
        self.label = label

    def __getitem__(self, index: int):
        """
        Returns the data and label of the index-th sample
        Args:
            index (int): Index of the required data point
        Returns:
            tuple: Binary data point of the index-th sample in the XOR dataset
            float: Label of the index-th sample in the XOR dataset
        """
        return self.data[index], self.label[index]

    def __len__(self):
        """Returns the size of the dataset

        Returns:
            int: Size of the dataset
        """
        return self.size
```

```python
# Creating the XOR dataset
dataset = XORdataset(size=10000, std_dev=0.1)
print("size of the dataset:", len(dataset))
print("first data point:", dataset[0][0])
```

# Optimization

Steps during training:  
    - Get a batch from the data loader  
    - Obtain the predictions from the model for the batch  
    - Calculate the loss based on the difference between predictions and labels  
    - Backpropagation: calculate the gradients for every parameter with respect to the loss  
    - Update the parameters of the model in the direction of the gradients


torch.optim has most popular optimizers implemented

```python
optimizer = torch.optim.SGD(myModel.parameters(), lr=0.1) # Stochastic Gradient Descent optimizer
```

The optimizer provides two useful functions: optimizer.step() updates the parameters based on the gradients, and optimizer.zero_grad() sets the gradients of all parameters to zero. While this function seems less relevant at first, it is a crucial pre-step before performing backpropagation.

If we call the "backward" function on the loss while the parameter gradients are non-zero from the previous batch, the new gradients would actually be added to the previous ones instead of overwriting them. This is done because a parameter might occur multiple times in a computation graph, and we need to sum the gradients in this case instead of replacing them. Hence, we need to call optimizer.zero_grad() before calculating the gradients of a batch.

There exist certain modules that need to perform a different forward step during training than during testing (e.g. BatchNorm and Dropout), and we can switch between them using model.train() and model.eval().

```python
def train_model(model, optimizer, data_loader, loss_module, num_epochs=50):
    '''
    Trains a model using the given optimizer, loss_module, and data_loader.
    Inputs:
        model: the model to train
        optimizer: the optimizer to use for training
        data_loader: the data loader providing the training data
        loss_module: the loss function to minimize
        num_epochs: the number of epochs to train
    '''
    model.train() # set the model to training mode
    for epoch in tqdm(range(num_epochs)):
        for data, label in data_loader:
            # Step 1: Move the data to the device (Only when GPU is available)
            data, label = data.to(device), label.to(device)
            
            # Step 2: Zero the gradients
            optimizer.zero_grad() # zero the gradients
            
            # Step 3: Run the model on input data
            preds = model(data) # forward pass
            preds = preds.squeeze(dim=1) # Output is [Batch size, 1], but we want [Batch size]
            
            # Step 4: Compute the loss
            loss = loss_module(preds, label.float())
            
            # Step 5: Perform backpropagation
            loss.backward()
            
            # Step 6: Update the parameters
            optimizer.step()
            
            
            
```

```python
train_model(myModel, optimizer, train_data_loader, loss_module)
```

# Evaluation

When evaluating the model, we don’t need to keep track of the computation graph as we don’t intend to calculate the gradients. This reduces the required memory and speed up the model. In PyTorch, we can deactivate the computation graph using with torch.no_grad(): .... Remember to additionally set the model to eval mode.

```python
test_data_loader = torch.utils.data.DataLoader(test_dataset, batch_size=128, shuffle=False, drop_last=False)

```

```python
def eval_model(model, data_loader):
    """
    Evaluates the model on the given data loader.
    Args:
        model (nn.Module): The model to evaluate
        data_loader (DataLoader): DataLoader providing the evaluation data
    Returns:
        float: The accuracy of the model on the data provided by the data loader
    """
    model.eval() # set the model to evaluation mode
    true_preds, num_preds = 0., 0.
    
    with torch.no_grad(): # disables gradient computation
        for data , label in data_loader:
            # Step 1: Move the data to the device (Only when GPU is available)
            data, label = data.to(device), label.to(device)
            
            # Step 2: Run the model on input data
            preds = model(data)
            preds = preds.squeeze(dim=1)
            preds = torch.sigmoid(preds) # convert logits to probabilities
            pred_labels = (preds > 0.5).to(torch.float32) # convert probabilities to binary predictions
            
            # Step 3: Record the true predictions and the number of predictions for computing the accuracy metric
            true_preds += (pred_labels == label).sum().item()
            num_preds += label.numel()
            
            # Step 4: Compute the accuracy
            acc = true_preds / num_preds # accuracy is the number of correct predictions divided by the number of total predictions
            print(f"Accuracy: {100.0*acc:4.2f}%")
            
```

# Visualizing Classification Boundaries

We can perform a prediction for every data point in a range of [-0.1, 1.5], and visualize the predicted class as in the sample figure at the beginning of this section. This shows where the model has created decision boundaries, and which points would be classified as 0, and which as 1.

```python
@torch.no_grad() # decorator that disables gradient computation for the function, same as using torch.no_grad() context manager
def visualize_classification(model, data, label):
    """
    Visualizes the classification results of the model on the given data.
    Args:
        model (nn.Module): The model to visualize
        data (Tensor): The input data
        label (Tensor): The ground truth labels
    """
    model.eval() # set the model to evaluation mode
    data, label = data.to(device), label.to(device)
    data_0 = data[label == 0]
    data_1 = data[label == 1]
    
    fig = plt.figure(figsize=(10, 5), dpi=300)
    plt.scatter(data_0[:, 0], data_0[:, 1], c='blue', label='Label 0')
    plt.scatter(data_1[:, 0], data_1[:, 1], c='red', label='Label 1')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.legend()
    
    # Visualize the decision boundary    
    model.to(device)
    c0 = torch.Tensor(to_rgba("C0")).to(device)
    c1 = torch.Tensor(to_rgba("C1")).to(device)
    x1 = torch.arange(-0.5, 1.5, step=0.01, device=device)
    x2 = torch.arange(-0.5, 1.5, step=0.01, device=device)
    xx1, xx2 = torch.meshgrid(x1, x2, indexing='ij') # creates a grid of points
    model_inputs = torch.stack([xx1, xx2], dim=-1)#.view(-1, 2)
    preds = model(model_inputs)
    preds = torch.sigmoid(preds)
    output_image = (1 - preds)*c0[None, None] + preds*c1[None, None] 
    output_image = output_image.numpy()
    plt.imshow(output_image, extent=(-0.5, 1.5, -0.5, 1.5), origin='lower')
    plt.grid(False)
    return fig
    
_ = visualize_classification(model=loaded_MyModel, data=dataset.data, label=dataset.label)
```

# TensorBoard logging

writer = SummaryWriter(...) to specify the directory in which the logging file should be saved, writer.add_.... to visualize things, for example, the computation graph with the function writer.add_graph, or a scalar value like the loss with writer.add_scalar. Finally, writer.add_figure() for visualization.

```python
# Import tensorboard logger from PyTorch
from torch.utils.tensorboard import SummaryWriter

# Load tensorboard extension for Jupyter Notebook, only need to start TB in the notebook
%reload_ext tensorboard
```

```python
def train_model_with_logger(model, optimizer, data_loader, loss_module, val_dataset, num_epochs=100, logging_dir='runs/our_experiment'):
    # Create TensorBoard logger
    writer = SummaryWriter(logging_dir)
    model_plotted = False

    # Set model to train mode
    model.train()

    # Training loop
    for epoch in tqdm(range(num_epochs)):
        epoch_loss = 0.0
        for data_inputs, data_labels in data_loader:

            ## Step 1: Move input data to device (only strictly necessary if we use GPU)
            data_inputs = data_inputs.to(device)
            data_labels = data_labels.to(device)

            # For the very first batch, we visualize the computation graph in TensorBoard
            if not model_plotted:
                writer.add_graph(model, data_inputs)
                model_plotted = True

            ## Step 2: Run the model on the input data
            preds = model(data_inputs)
            preds = preds.squeeze(dim=1) # Output is [Batch size, 1], but we want [Batch size]

            ## Step 3: Calculate the loss
            loss = loss_module(preds, data_labels.float())

            ## Step 4: Perform backpropagation
            # Before calculating the gradients, we need to ensure that they are all zero.
            # The gradients would not be overwritten, but actually added to the existing ones.
            optimizer.zero_grad()
            # Perform backpropagation
            loss.backward()

            ## Step 5: Update the parameters
            optimizer.step()

            ## Step 6: Take the running average of the loss
            epoch_loss += loss.item()

        # Add average loss to TensorBoard
        epoch_loss /= len(data_loader)
        writer.add_scalar('training_loss',
                          epoch_loss,
                          global_step = epoch + 1)

        # Visualize prediction and add figure to TensorBoard
        # Since matplotlib figures can be slow in rendering, we only do it every 10th epoch
        if (epoch + 1) % 10 == 0:
            fig = visualize_classification(model, val_dataset.data, val_dataset.label)
            writer.add_figure('predictions',
                              fig,
                              global_step = epoch + 1)

    writer.close()
```

```python
model = simpleClassifier(input_size=2, hidden_size=4, num_classes=1).to(device)
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
train_model_with_logger(model, optimizer, train_data_loader, loss_module, val_dataset=dataset)
```
